```python
#correlation analysis
numeric_df = df.select_dtypes(include=[np.number])
```

```python
# Histograms for selected numeric columns
for col in ['ev_sales', 'petrol_car_sales', 'diesel_car_sales', 'total_vehicle_sales', 'ev_market_share']:
    if col in df.columns:
        plt.figure()
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.tight_layout()
        plt.show()
```

```python
# Pie Chart or count plot for a categorical variable 'region'
if 'region' in df.columns:
    plt.figure()
    sns.countplot(data=df, x='region')
    plt.title('Count Plot of Regions')
    plt.xlabel('Region')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Box Plot to inspect distribution of EV market share
if 'ev_market_share' in df.columns:
    plt.figure()
    sns.boxplot(x=df['ev_market_share'])
    plt.title('Box Plot of EV Market Share')
    plt.tight_layout()
    plt.show()
```

```python
#Global Sales Trend (Yearly)
yearly_sales = df.groupby("year")[["ev_sales", "petrol_car_sales", "diesel_car_sales"]].sum()

plt.figure(figsize=(12,6))
yearly_sales.plot()
plt.title("EV vs Petrol vs Diesel Sales Over Time")
plt.ylabel("Sales")
plt.show()
```

```python
#EV Market Share Trend
ev_share_year = df.groupby("year")["ev_market_share"].mean()

plt.figure(figsize=(10,5))
ev_share_year.plot(marker="o", color="green")
plt.title("Average EV Market Share Over Time")
plt.ylabel("EV Market Share (%)")
plt.show()
```

```python
#Country-Level EV Dominance
dominant_countries = df[df["is_ev_dominant"] == 1]["country"].unique()
print("Countries where EV became dominant:")
print(dominant_countries)
```

```python
# correlation heatmap if we have four or more numeric columns
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(12, 10))
    correlation_matrix = numeric_df.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.tight_layout()
    plt.show()
```
